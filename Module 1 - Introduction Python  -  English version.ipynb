## 1.- Basic Concepts

Python is a high-level language, which means that the operating system will not execute our program instructions directly.

In order for a program written in a high-level language to be executed, it must first be compiled or interpreted. In our case, the Pyton Kernnel will interpret the instructions so that they can be executed.



When a program written in a high-level language is compiled, a file with an executable file extension is generated (for example, in Windows .exe). This file can be executed by any computer that has the same operating system. Although it does not have the compiler, since the instructions are specific to the processor.

When it is interpreted, that file is not constructed, so it can only be executed if the interpreter is installed.
