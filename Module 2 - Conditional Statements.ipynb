## 1. Control Flow

In Python, instructions are executed sequentially (from top to bottom)

The <span style="color:red">Flow Control</span> instructions allow you to BREAK this sequence. There are two types of control statements:

### 1.A. Conditional Sentences

     Blocks of statements that are executed depending on the result of a boolean expression (a condition)
    
### 1 B. Repetitive Sentences (loops)

     Blocks of statements that repeat as long as a condition is met
    
  <span style="color:green"> An algorithm should be written to be executed sequentially (by default) and/or using control flow statements: Conditional Statements and/or Repetitive Statements </span>

  In this module we will focus on CONDITIONAL Statements

## 2. Conditional Statement <span style="color:red"> if ELSE </span>

This is one of the most basic conditional statements.

In Flow Diagrams, the conditional statement has similar structure to the following:



##### <span style="color:green"> If the condition is met /**True**, the statement/block of statments associated with **if** is executed.
    
##### <span style="color:orange"> If the condition is not met/**False**, the statement/block of statments associated with **else** is executed.
   
### <span style="color:red"> Practical Exercise </span>
Run the program in the following cell for different entries (enter the letter "a" first then try other entries (not "a")).

```python
#cell 1

letter = input("Enter a letter")
if letter == "a":
     print ("Okay, you have typed the letter 'a'") #This statement is executed only if the condition is True (the user entres the letter a)
else:
     print ("You have typed something other than 'a'") #This statement is executed only if the condition is False
print ("Whether you typed 'a' or something else, this print is always executed.") #This statement is always executed
```

### <span style="color:green">Details to take into account:</span>
#### <span style="color:green">1. Syntax:</span>


#### <span style="color:green">2. Boolean condition:</span>
When you write a conditional statement, you will always have the word <span style="color:green">"**if**"</span> and the sign "**:**", indicating a **logical condition**.

The condition can be simple or a complex expression, but the expression of the logical condition must ALWAYS be RESULT IN boolean: <span style="color:green">**True**</span> ( true) or <span style="color:green">**False**</span> (False).

#### <span style="color:green">3. Indentation/Tabulation in statements of the same BLOCK:</span>
After you type ":" and press ENTER, the editor automatically indents/tabs to the next line, moving it a few spaces to the right. This detail seems insignificant, but IT IS NOT. BE VERY CAREFUL WITH THIS IN PYTHON:

**All statements belonging to the same BLOCK in Python MUST HAVE THE SAME INDENTION**

That is, in this case, as long as you write new statements shifted to the right, they will be part of the <span style="color:green">GREEN</span> block of the graph (belonging to the IF component of the statement). All those instructions are executed sequentially if the condition is met/True.

Other things to be taken into account when writting a block:
1) **The NUMBER and TYPE of statements** of the block is unlimited**
2) **There must be at least ONE STATEMENT in each block** of statements. Although Python has a statement to indicate that nothing needs to be execute, it is called **pass**, a good programmer would never declare blocks if it has no statements to be execute, there wont be any use of it.  

When you're done with that block of statements, on the next line, press delete key **(only once)** and the indentation will end. Now you can write the phrase <span style="color:green">**else :**</span>

<i> Again the editor helps you with the indentation </i>. Write the instructions in the orange block, they will only be executed if the condition is **False**

When you finish writing the <span style="color:green">**else**</span> statement, on the next line, press the delete key **(only once)** and it will finish the indentation (i.e. the else block of statements).

Other programming languages delimit the statements of a block by encapsulating special signed statements, for example, {}. In Python they are identified with : and with indentation. Therefore, although indentation is optional and highly recommended in many languages, in Python it is mandatory.
